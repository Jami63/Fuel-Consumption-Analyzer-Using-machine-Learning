```python
import pandas as P
```

```python
D = P.read_csv("Car Data.csv")
```

```python
import matplotlib.pyplot as PLOT
```

```python
print(D.columns)
```

```python
PLOT.hist (D["combination_mpg"] , bins = 25)
PLOT.xlabel("Combination MPG")
PLOT.ylabel("Frequency")
PLOT.figure ()
PLOT.show()
```

```python
PLOT.scatter(D["city_mpg"] , D["highway_mpg"])
PLOT.xlabel("City MPG")
PLOT.ylabel("Highway MPG")
PLOT.figure ()
PLOT.show()
```

```python
PLOT.scatter(D["combination_mpg"] , D["displacement"])
PLOT.xlabel("Combination MPG")
PLOT.ylabel("Engine Displacement")
PLOT.figure ()
PLOT.show()
```

```python
FT_MPG = D.groupby("fuel_type")["combination_mpg"].mean()
```

```python
FT_MPG.plot(kind = "bar")
PLOT.xlabel("Fuel Type")
PLOT.ylabel("Average Combination MPG")
PLOT.figure ()
PLOT.show()
```
