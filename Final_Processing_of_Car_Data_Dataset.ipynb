```python
N = D.select_dtypes(include='number')
```

```python
Q1 = N.quantile(0.25)
Q2 = N.quantile(0.75)
IQR = Q2 - Q1
```

```python
print(Q1)
print("\n")
print(Q2)
print("\n")
print(IQR)
```

```python
Outlier = (N < (Q1 - 1.5 * IQR)) | (N > (Q2 + 1.5 * IQR))
print(Outlier)
print(Outlier.sum())
```

```python
L = Q1 - 1.5 * IQR
U = Q2 + 1.5 * IQR
D = N.clip(lower = L, upper = U, axis=1)
```
