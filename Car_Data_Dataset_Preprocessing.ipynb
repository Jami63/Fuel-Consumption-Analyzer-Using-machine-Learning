Check Dataset Outlier

```python
Q1 = D.quantile(0.25)
Q3 = D.quantile(0.75)
IQR = Q3 - Q3
```

Removing Outlier

```python
Outlier = (D < (Q1 - 1.5 * IQR)) | (D > (Q3 + 1.5 * IQR))
print(Outlier)
print(Outlier.sum())
```

```python
L = Q1 - 1.5 * IQR
U = Q3 + 1.5 * IQR
D = D.clip(lower = L, upper = U, axis=1)
```

Check Outlier Again

```python
print(L)
print("\n")
print(U)
print("\n")
print(D)
```
